```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('SSB150-150.5MHZ.npy')
```

```python
data = data[0]
data = data - data.mean()

data_fft = abs(np.fft.fft(data))**2
    
data_fft = np.fft.fftshift(data_fft)
    
print(len(data))
time=np.arange(0,2048/2.2e6,1/2.2e6)
```

```python
plt.plot(time, data_fft)
plt.show()
```

```python
sindata= np.load('sin500.npy')
sindata = sindata[0]
sindata = sindata - sindata.mean()

sindata_fft = abs(np.fft.fft(sindata))**2
    
sindata_fft = np.fft.fftshift(sindata_fft)
plt.plot(time, data_fft, label = 'data', linewidth = 3)
plt.plot(time, sindata_fft, label = 'sin wave', linewidth = 0.5)
plt.legend()
plt.xlim(0.0002, 0.0008)
plt.ylim(0,40)
plt.show()
```

```python
fft_freq = np.fft.fftshift(np.fft.fftfreq(2048,1/2.2))
```

```python
plt.plot(fft_freq, data_fft, label = '150.5 MHz', linewidth = 3)
plt.plot(fft_freq, sindata_fft, label = '500 khzsin wave', linewidth = 0.5)
plt.legend()
#plt.xlim(-0.6, 0.6)
plt.ylim(0,100)
plt.xlabel('MHz')
plt.ylabel('Power (arbitrary)')
plt.show()
```

```python
data_neg = np.load('SSB150-149.5MHZ.npy')
data_neg = data_neg[0]
data_neg = data_neg - data_neg.mean()

data_neg_fft = abs(np.fft.fft(data_neg))**2
    
data_neg_fft = np.fft.fftshift(data_neg_fft)
```

```python
plt.plot(fft_freq, data_fft, label = '150.5 MHz', linewidth = 3)
plt.plot(fft_freq, data_neg_fft, label = '149.5 MHz', linewidth = 3)
plt.plot(fft_freq, sindata_fft, label = '0.5 MHz sin wave', linewidth = 1.5)
plt.legend()
#plt.xlim(-0.6, 0.6)
plt.ylim(0,1000)
plt.xlabel('MHz')
plt.ylabel('Power (arbitrary)')
plt.show()
```
