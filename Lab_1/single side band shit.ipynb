```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('SSB30MHZ-30.5MHZ.im[phaseshift].npy')
```

```python
data = data[0]
data = data - data.mean()

data_fft = abs(np.fft.fft(data))
    
data_fft = np.fft.fftshift(data_fft)
    
print(len(data))
time=np.arange(0,2048/2.2e6,1/2.2e6)
```

```python
plt.plot(time, data_fft)
plt.show()
```

```python
sindata= np.load('sin500.npy')
sindata = sindata[0]
sindata = sindata - sindata.mean()

sindata_fft = abs(np.fft.fft(sindata))
    
sindata_fft = np.fft.fftshift(sindata_fft)
plt.plot(time, data_fft, name = 'data')
plt.plot(time, sindata_fft, name = 'sin wave')
plt.legend()
plt.show()
```
