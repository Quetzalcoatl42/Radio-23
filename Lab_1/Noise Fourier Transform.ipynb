```python
data1 = np.load('noise1.npy')
data2 = np.load('noise2.npy')
data3 = np.load('noise3.npy')
data4 = np.load('noise4.npy')
data5 = np.load('noise5.npy')
```

```python
np.concatenate((data1, data2, data3, data4, data5))
```

```python
noise_data_full = np.concatenate((data1, data2, data3, data4, data5))
```

```python
noise_data_full = noise_data_full - noise_data_full.mean()
data_fft = abs(np.fft.fft(noise_data_full)) #Fourier Transform
data_fft = np.fft.fftshift(data_fft)
data = data_fft[0]
"""#Test Data
print(data_fft[0][100])
print(data_fft.real[0][100])
print(data_fft.imag[0][100])
"""

#Build Plot
#Frequency=np.arange(?????????????????????????)
time=np.arange(0,2048/2.2e6,1/2.2e6) #3.2e6 Sample Rate CHA


plt.plot(time,data) #FFT

#plt.plot(time,data_fft[0]**2) #Power Spectrum



#plt.legend()
plt.title('My title')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Volts??? [V]')

plt.show()
```
