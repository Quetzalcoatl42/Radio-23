```python

import numpy as np
import matplotlib.pyplot as plt

gs=['sin300.npy','sin500.npy','sin700.npy','sin900.npy', 'sin1100.npy', 'sin1300.npy', 'sin1500.npy']
for g in gs:

    data = np.load(g) #Insert Data File Name
    data = data - data.mean()
    
    data_fft = abs(np.fft.fft(data)) #Fourier Transform
    #data_fft = np.fft.fftshift(data_fft)
    
    #Test Data
    #print(data_fft[0][100])
    #print(data_fft.real[0][100])
    #print(data_fft.imag[0][100])
    
    
    #Build Plot
    #Frequency=np.arange(?????????????????????????)
    time=np.arange(0,2048/2.2e6,1/2.2e6) #3.2e6 Sample Rate CHA
    
    if g == 'sin300.npy':
        labels = '300 KHz'
        width = 1.5
    elif g == 'sin500.npy':
        labels = '500 KHz'
        width = 1.5
    elif g == 'sin700.npy':
        labels = '700 KHz'
        width = 1
    elif g == 'sin900.npy':
        labels = '900 KHz'
        width = 1
    elif g == 'sin1100.npy':
        labels = '1100 KHz'
        width = 1.5
    elif g == 'sin1300.npy':
        labels = '1300 KHz'
        width = 3
    elif g == 'sin1500.npy':
        labels = '1500 KHz'
        width = 3
    
    
    plt.plot(time,data_fft[0], label = labels, linewidth = width) #FFT
    
    #plt.plot(time,data_fft[0]**2) #Power Spectrum
    
    

plt.legend()
plt.title('Measured sin wave Fourier transforms')
plt.ylim(0,250)
#plt.xlabel('Frequency [Hz]')
#plt.ylabel('Volts??? [V]')

plt.show()
```
