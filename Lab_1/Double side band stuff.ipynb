```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('DSB600-630.npy')
```

```python
data = data[0]
data = data - data.mean()

data_fft = abs(np.fft.fft(data))
    
data_fft = np.fft.fftshift(data_fft)
    
print(len(data))
time=np.arange(0,2048/2.2e6,1/2.2e6)
```

```python
plt.plot(time, data_fft)
plt.show()
```

```python
data_minus = np.load('DSB600-570.npy')
data_minus = data_minus[0]
data_minus = data_minus - data_minus.mean()

data_minus_fft = abs(np.fft.fft(data_minus))
    
data_minus_fft = np.fft.fftshift(data_minus_fft)
    
print(len(data_minus))
time=np.arange(0,2048/2.2e6,1/2.2e6)
```

```python
plt.plot(time, data_fft)
plt.plot(time, data_minus_fft)
#plt.xlim(0.0004,0.0006)
plt.show()
```

```python
data_30 = np.load('sin030.npy')
data_1170 = np.load('sin1170.npy')
data_1230 = np.load('sin1230.npy')
```

```python
data_30 = data_30[0]
data_30 = data_30 - data_30.mean()

data_30_fft = abs(np.fft.fft(data_30))
    
data_30_fft = np.fft.fftshift(data_30_fft)
data_1170 = data_1170[0]
data_1170 = data_1170 - data_1170.mean()

data_1170_fft = abs(np.fft.fft(data_1170))
    
data_1170_fft = np.fft.fftshift(data_1170_fft)
data_1230 = data_1230[0]
data_1230 = data_1230 - data_1230.mean()

data_1230_fft = abs(np.fft.fft(data_1230))
    
data_1230_fft = np.fft.fftshift(data_1230_fft)
```

```python
plt.plot(time, data_30_fft, linewidth = 0.5, label = '30 kHz')
plt.plot(time, data_1170_fft, linewidth = 0.5, label = '1170 kHz')
plt.plot(time, data_1230_fft, linewidth = 0.5, label = '1230 kHz')
plt.plot(time, data_fft, label = 'Radio Frequency 630 kHz')
plt.plot(time, data_minus_fft, label = 'Radio Frequency 570 kHz')
#plt.xlim(0.0004,0.0006)
plt.legend()
plt.show()
```

```python
plt.plot(time, data_30_fft, linewidth = 0.5, label = '30 kHz')
plt.plot(time, data_1170_fft, linewidth = 0.5, label = '1170 kHz')
plt.plot(time, data_1230_fft, linewidth = 0.5, label = '1230 kHz')
plt.legend()
plt.show()
```
