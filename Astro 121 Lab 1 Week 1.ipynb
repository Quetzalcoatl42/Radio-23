```python
x = np.arange(0,10*np.pi,0.01)
y = np.sin(x) + np.sin(2*x/3)
y1= np.fft.fft(y)
y2= np.absolute(y1)**2
plt.figure(figsize=(10,5))
plt.xlim(0,10*np.pi)
plt.title('Sine Wave and its Fourier Transform')
plt.plot(x,y,y2/np.max(y2))
plt.show
```
