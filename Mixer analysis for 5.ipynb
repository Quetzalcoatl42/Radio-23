```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
for filename in os.listdir('mix_files'):
    print(filename)
```

```python
for mix in os.listdir('mix_files'):
    data = np.load('mix_files/' + str(mix))

    data_fft = abs(np.fft.fft(data))
    print(len(data))
    time=np.arange(0,2048/2.2e6,1/2.2e6) #3.2e6 Sample Rate CHA


    plt.plot(time,data_fft[0]) #FFT

#plt.plot(time,data_fft[0]**2) #Power Spectrum



    plt.legend()
    plt.title(str(mix))
    plt.xlabel('Frequency [Hz]')
    plt.xlim(0.00062, 0.00065)
    plt.ylabel('Volts??? [V]')

    plt.show()
```

```python
for mix in os.listdir('mix_files'):
    data = np.load('mix_files/' + str(mix))
    data = data[0][:200]
    data_fft = abs(np.fft.fft(data))
    print(len(data))
    time=np.arange(0,2048/2.2e6,1/2.2e6) #3.2e6 Sample Rate CHA


    plt.plot(time[:200],data_fft[0]) #FFT

#plt.plot(time,data_fft[0]**2) #Power Spectrum



    plt.legend()
    plt.title(str(mix))
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Volts??? [V]')

    plt.show()
```
